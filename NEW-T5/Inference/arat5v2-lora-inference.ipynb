```python
from peft import PeftModel, PeftConfig
from transformers import AutoModelForSeq2SeqLM
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("UBC-NLP/AraT5v2-base-1024")

config = PeftConfig.from_pretrained("yasmineee/araT5-Base-with-LoRA")
base_model = AutoModelForSeq2SeqLM.from_pretrained("UBC-NLP/AraT5v2-base-1024")
model = PeftModel.from_pretrained(base_model, "yasmineee/araT5-Base-with-LoRA")


def translate(text, model, tokenizer, max_length=50, num_beams=5):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=256)
    outputs = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_length=max_length,
        num_beams=num_beams,
        length_penalty=1.0,
        early_stopping=True,
        do_sample=True,
        top_k=120,
        top_p=0.95
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example usage
text = "translate Arabic to English: بما في ذلك التعليم، والصحة، والتكنولوجيا."
translated_text = translate(text, model, tokenizer)
print("Translation:", translated_text)
```
